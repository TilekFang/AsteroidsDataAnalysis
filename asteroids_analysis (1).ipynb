```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('nasa.csv')
```

```python
df.head()
```

### Exploring Data

```python
df.info()
```

### Sorting elements to keep only Hazardous for Earth:

```python
dang_ast = df[df['Hazardous'] == True]
```

```python
dang_ast['Hazardous'].head()
```

```python
dang_ast.describe()
```

### Dropping excess elements:

```python
dang_ast['Equinox'].unique().sum()
```

```python
dang_ast = dang_ast.drop(['Hazardous', 'Equinox', 'Est Dia in Feet(max)', 'Miles per hour', 'Miss Dist.(Astronomical)', 'Miss Dist.(lunar)', 'Miss Dist.(miles)'], axis = 1)
```

```python
dang_ast.head()
```

```python
dang_ast.columns
```

```python
dang_ast.drop(['Relative Velocity km per sec'], axis = 1)
```

```python
dang_ast = dang_ast.drop(['Est Dia in KM(min)', 'Est Dia in KM(max)'], axis = 1)
```

```python
dang_ast = dang_ast.drop(['Epoch Date Close Approach', 'Relative Velocity km per sec', 'Orbiting Body', ], axis = 1)
```

```python
dang_ast[['Perihelion Distance',
       'Perihelion Arg', 'Aphelion Dist', 'Perihelion Time']]
```

### Exploring duplicates and dropping it:

```python
dang_ast.duplicated('Name').sum()
```

```python
dang_ast = dang_ast.drop_duplicates(subset = ['Name'])
```

```python
dang_ast.duplicated('Name').sum()
```

### Sorting values by distance in Ascending order:

```python
ast_dist = dang_ast.sort_values(by = ['Miss Dist.(kilometers)'])
```

```python
ast_dist.columns
```

```python
ast_dist = ast_dist.drop(['Est Dia in Miles(min)', 'Est Dia in Miles(max)',
       'Est Dia in Feet(min)'], axis = 1)
```

```python
ast_dist[ast_dist['Miss Dist.(kilometers)'] < 1.923362e+07]
```

### Creating scatter plot to see the magnitude of asteroids per distance objects and Earth:

```python
plt.scatter(ast_dist['Absolute Magnitude'], ast_dist['Miss Dist.(kilometers)'], c = 'b')
plt.xlabel('The Magnitude of Asteroids')
plt.ylabel('The distance between Earth and Asteroids')
plt.show()
```

```python
ast_dist[ast_dist['Absolute Magnitude'] == 14.400000]
```

```python
#less_than_half = ast_dist[ast_dist['Miss Dist.(kilometers)'] < 3.889844e+07]
ast_dist['Miss Dist.(kilometers)'].describe()
```

```python
ast_dist['Miss Dist.(kilometers)'].min()
```

### Converting date from 'object' to date type:

```python
ast_dist['Close Approach Date']
```

```python
ast_dist['Close Approach Date'] = pd.to_datetime(ast_dist['Close Approach Date'])
```

```python
ast_dist['Orbit Determination Date'] = pd.to_datetime(ast_dist['Orbit Determination Date'])
```

```python
ast_dist['Relative Velocity km per hr'].describe()
```

### Exploring the speed by histogram:

```python
plt.hist(ast_dist['Relative Velocity km per hr'], bins=40, density=True)
plt.title('Relative Velocity', fontsize = 13)
plt.xlabel('The speed in km/h')
plt.show()
```

```python
ast_dist['Est Dia in M(avg)'] = (ast_dist['Est Dia in M(min)'] + ast_dist['Est Dia in M(max)'])/2
```

```python
n = ast_dist['Est Dia in M(avg)'].round(decimals = 2)
n
```

### Calculating the size of asteroids according to following formula:
### Size (in meters) = (Diameter (in meters) / 2)^2 * Ï€

```python
ast_dist['Size(meters)'] = (n / 2) ** 2 * 3.14
ast_dist['Size(meters)']
```

```python
ast_dist['Size(meters)'].describe()
```

### Plotting Size column to see the most common size of asteroids:

```python
plt.eventplot(ast_dist['Size(meters)'], color = 'purple')
plt.title('The sizes of objects')
plt.show()
```

### Creating 3D sctter plot to see the relationalship between 3 columns:
